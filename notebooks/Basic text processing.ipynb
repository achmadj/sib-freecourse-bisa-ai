```python
import pandas as pd
import re, string

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

from tqdm.auto import tqdm

list_stopwords = stopwords.words("english")
```

```python
with open("./sentiment labelled sentences/amazon_cells_labelled.csv", "r") as f:
  fsr = f.readlines()

data = []

for text in tqdm(fsr):
  strin = re.findall(r"[^,]+", text)
  s = " ".join(strin)
  s = s[:-4] + "," + s[-3:]
  data.append(s)

with open("data.csv", "w+") as f:
  f.write("".join(data))
```

```python
df = pd.read_csv("./data.csv", header=None)
df.columns = ["text", "class"]
df
```

## Case Folding dan Removing number

```python
df["text"] = df["text"].str.lower().str.replace("\d", "", regex=True).str.replace(f"[{string.punctuation}]", "", regex=True)
df
```

## Tokenization

```python
tokens = []

for text in tqdm(df.text):
  tokens.append(word_tokenize(text))

df["text"] = tokens
df
```

## Removing StopWords

```python
df["text"] = df["text"].apply(lambda x: " ".join([word for word in x if word not in (list_stopwords)]))
df
```
